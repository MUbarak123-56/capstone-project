```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/resources/data/Topic_Survey_Assignment.csv",  index_col = False)
```

```python
df.head()
```

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df.sort_values(['Very interested'], ascending = False, axis = 0, inplace = True) 
df.rename(columns={'Unnamed: 0': 'Topic'}, inplace = True)
red = df.set_index('Topic', inplace= True)
```

```python
b = df.plot(kind='bar',figsize=(20, 8), color = ('#5cb85c','#5bc0de','#d9534f'))
b
plt.title("Respondents' Interests in Data Science Areas", fontsize=16)
plt.xticks(red, fontsize = 14) 
b.spines['top'].set_visible(False)
b.spines['right'].set_visible(False)
b.spines['bottom'].set_linewidth(0.5)
b.spines['left'].set_visible(True)
plt.legend(loc="upper right", fontsize=14)
plt.show()
```
